1. Data Cleaning

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '/content/IMB469-XLS-ENG (1).xlsx'
df = pd.read_excel(file_path, sheet_name='IMB 469 Auto Finance')

# Convert Start_Date to datetime and handle missing values
df['Start_Date'] = pd.to_datetime(df['Start_Date'], errors='coerce')
df['Start_Date'] = df['Start_Date'].fillna(pd.Timestamp('1900-01-01'))

# Fill or drop other missing values as needed
df.fillna(0, inplace=True)  # Example: fill numeric columns with 0
```

2. Outliers by Percentile Method

```python
# Define a function to remove outliers using percentiles
def remove_outliers(df, columns, lower_percentile=0.01, upper_percentile=0.99):
    for col in columns:
        lower = df[col].quantile(lower_percentile)
        upper = df[col].quantile(upper_percentile)
        df = df[(df[col] >= lower) & (df[col] <= upper)]
    return df

numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
df = remove_outliers(df, numeric_cols)
```

3.Exploratory Data Analysis

```python
# Summary statistics
print("Summary Statistics:")
print(df.describe())

# Basic information about the dataset
print("\nBasic Information:")
print(df.info())

# Box Plot for Exploratory Data Analysis
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.select_dtypes(include=['float64', 'int64']))
plt.title("Box Plot for Numerical Variables")
plt.xticks(rotation=45)
plt.show()

```

```python
# Assuming `MTHINCTH` represents income in thousands
income_column = 'MTHINCTH'

# Calculate percentiles
percentiles = df[income_column].quantile([0.25, 0.5, 0.75, 1.0])  # 25th, 50th, 75th, and max percentiles
print("Percentiles:")
print(percentiles)

# Create frequency distribution
frequency_distribution = df[income_column].value_counts(bins=5, sort=False).reset_index()
frequency_distribution.columns = ['Income Range', 'Frequency']

# Calculate cumulative frequency and percentage
frequency_distribution['Cumulative Frequency'] = frequency_distribution['Frequency'].cumsum()
frequency_distribution['Percentage'] = (frequency_distribution['Frequency'] / len(df)) * 100
frequency_distribution['Cumulative Percentage'] = frequency_distribution['Percentage'].cumsum()

# Display the frequency distribution table
print("\nFrequency Distribution with Cumulative Frequency and Percentage:")
print(frequency_distribution)

# Plot Frequency Distribution
import matplotlib.pyplot as plt
frequency_distribution.set_index('Income Range')['Frequency'].plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title("Income Frequency Distribution")
plt.xlabel("Income Range (in Thousands)")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

5. Variance Inflation Factor (VIF)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# Calculate VIF
X = add_constant(df[numeric_cols])
vif = pd.DataFrame()
vif["variables"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
print(vif)
```

7. Binning

```python
# Create bins for AGE
df['AGE_Bin'] = pd.cut(df['AGE'], bins=[0, 25, 35, 50, 100], labels=['Young', 'Mid-age', 'Adult', 'Senior'])

# Create bins for MTHINCTH (monthly income)
df['Income_Bin'] = pd.cut(df['MTHINCTH'], bins=[0, 5, 10, 20, 50], labels=['Low', 'Lower-Middle', 'Upper-Middle', 'High'])

# Cross-tabulation to analyze defaults
default_analysis = pd.crosstab(index=df['AGE_Bin'],
                                columns=df['Income_Bin'],
                                values=df['DefaulterFlag'],
                                aggfunc='mean')

# Display the result
print("Default Rates by Age and Income Category:")
print(default_analysis)

# Identify the category with the highest default rate
highest_default = default_analysis.stack().idxmax()
print(f"The category with the highest default rate is: {highest_default}")

```

8. Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Define features and target
X = df[['AGE', 'MTHINCTH', 'DWNPMFR']]  # Example features
y = df['DefaulterFlag']  # Target variable

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report

# Define features and target variable
X = df[['AGE', 'MTHINCTH', 'DWNPMFR']]  # Example features
y = df['DefaulterFlag']  # Target variable (0: No Default, 1: Default)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=model.classes_)
disp.plot(cmap="Blues")
plt.title("Confusion Matrix")
plt.show()

# Classification report for additional metrics
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Generate classification report
report = classification_report(y_test, y_pred)
print("\nClassification Report:")
print(report)

```
